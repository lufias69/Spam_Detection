```python
import pandas as pd

from chi import square as cs
sf = cs.ChiSquare(alpha = 0.1)
```

```python
data = pd.read_excel("komentar/data_clean.xlsx")
komentar = data.komentar.tolist()
label = data.label.tolist()
data.head()
```

```python
sf.find_best_features(komentar, label, fitur = None)
```
